```python
import pandas as pd
from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler
import matplotlib.pyplot as plt
```

```python
from google.colab import drive
drive.mount('/content/drive')
```

```python
data = pd.read_csv("/content/drive/MyDrive/Mall_Customers.csv")
features = data[[ 'Annual Income (k$)','Spending Score (1-100)']]
```

```python
scaler = StandardScaler()
scaled_features = scaler.fit_transform(features)
```

```python
kmeans = KMeans(n_clusters=3, random_state=42)
kmeans.fit(scaled_features)
data["cluster"] = kmeans.labels_
centroids = scaler.inverse_transform(kmeans.cluster_centers_)
```

```python
plt.figure(figsize=(8, 6))
for cluster_id in range(3):
    plt.scatter(
        data[data["cluster"] == cluster_id]["Annual Income (k$)"],
        data[data["cluster"] == cluster_id]["Spending Score (1-100)"],
        label=f"Cluster {cluster_id}",
        cmap="viridis",
        alpha=0.7,
    )

# Plot centroids
plt.scatter(centroids[:, 0], centroids[:, 1], marker="X", color="red", s=100, label="Centroids")

plt.xlabel("Annual Income")
plt.ylabel("Spending Score")
plt.title("Customer Segmentation with Centroids")
plt.colorbar(label="Cluster")
plt.legend()
plt.show()
```
